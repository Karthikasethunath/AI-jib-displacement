```python
import logging
from linkedin_jobs_scraper import LinkedinScraper
from linkedin_jobs_scraper.events import Events, EventData, EventMetrics
from linkedin_jobs_scraper.query import Query, QueryOptions, QueryFilters
from linkedin_jobs_scraper.filters import RelevanceFilters, TimeFilters, TypeFilters, ExperienceLevelFilters, \
    OnSiteOrRemoteFilters
import csv

# Change root logger level (default is WARN)
logging.basicConfig(level=logging.INFO)

# Create a list to store scraped job data
job_data_list = []

# Fired once for each successfully processed job
def on_data(data: EventData):
    job_data = {
        "Title": data.title,
        "Company": data.company,
        "Company Link": data.company_link,
        "Date": data.date,
        "Link": data.link,
        "Insights": data.insights,
        "Description Length": len(data.description)
    }
    job_data_list.append(job_data)

# Fired once for each page (25 jobs)
def on_metrics(metrics: EventMetrics):
    print('[ON_METRICS]', str(metrics))

def on_error(error):
    print('[ON_ERROR]', error)

def on_end():
    print('[ON_END]')

    # Write the scraped job data to a CSV file
    with open('job_data.csv', 'w', newline='') as csv_file:
        fieldnames = ["Title", "Company", "Company Link", "Date", "Link", "Insights", "Description Length"]
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
        for job_data in job_data_list:
            writer.writerow(job_data)

scraper = LinkedinScraper(
    chrome_executable_path=None,
    chrome_options=None,
    headless=True,
    max_workers=1,
    slow_mo=0.5,
    page_load_timeout=40
)

scraper.on(Events.DATA, on_data)
scraper.on(Events.ERROR, on_error)
scraper.on(Events.END, on_end)

queries = [
    Query(
        options=QueryOptions(
            limit=27
        )
    ),
    Query(
        query='Engineer',
        options=QueryOptions(
            locations=['United States', 'Europe'],
            apply_link=True,
            skip_promoted_jobs=True,
            page_offset=2,
            limit=5,
            filters=QueryFilters(
                company_jobs_url='https://www.linkedin.com/jobs/search/?f_C=1441%2C17876832%2C791962%2C2374003%2C18950635%2C16140%2C10440912&geoId=92000000',
                relevance=RelevanceFilters.RECENT,
                time=TimeFilters.MONTH,
                type=[TypeFilters.FULL_TIME, TypeFilters.INTERNSHIP],
                on_site_or_remote=[OnSiteOrRemoteFilters.REMOTE],
                experience=[ExperienceLevelFilters.MID_SENIOR]
            )
        )
    ),
]

scraper.run(queries)
```

```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the LinkedIn search page
url = 'https://www.linkedin.com/search/results/people/?keywords=data%20scientist'
headers = {
    'User-Agent': 'Your User-Agent String',
}

response = requests.get(url, headers=headers)

# Parse the HTML response
soup = BeautifulSoup(response.text, 'html.parser')

# Find and extract usernames
usernames = []
for profile_link in soup.find_all('a', class_='search-result__result-link'):
    username = profile_link.get('href').split('/')[-2]  # Extract the username from the URL
    usernames.append(username)

# Print the list of usernames
print(usernames)
```
